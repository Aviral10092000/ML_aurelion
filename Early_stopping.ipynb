```python
df.drop("pm25tmean2",axis=1,inplace=True)
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
df.drop("date",axis=1,inplace=True)
```

```python
df.info()
```

```python
float_pipeline = []
object_pipeline = []
index = 0
for _types in df.dtypes:
    if _types == "object":
        object_pipeline.append(df.columns[index])
        pass
    else:
        float_pipeline.append(df.columns[index])
        pass
    index = index + 1
    pass
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

num_pipline = Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    ('std_scaler',StandardScaler())
])

full_pipeline = ColumnTransformer([
    ("num",num_pipline,float_pipeline),
    ("cat",OneHotEncoder(),object_pipeline)
])
```

```python
prep = full_pipeline.fit_transform(df)
```

```python
prep.toarray()
```

```python
prep[:,:-1]
```

```python
from sklearn.base import clone
from sklearn.linear_model import SGDRegressor
import numpy as np
from sklearn.metrics import mean_squared_error


sgd_reg = SGDRegressor(max_iter=1,tol=-np.infty,warm_start=True,penalty=None,learning_rate="constant",eta0=0.0005)
```

```python
minimum_val_error = float("inf")
best_epoch = None
best_model = None
for epoch in range(1000):
    sgd_reg.fit(prep[:,:-1],prep[:,-1])
    y_val_predict = sgd_reg.predict(prep[:,:-1])
    val_error = mean_squared_error(y_val_predict,prep[:,-1])
    if val_error < minimum_val_error:
        minimum_val_error = val_error
        best_epoch = epoch
        best_model = clone(sgd_reg)
        pass
```
