```python
import pandas as pd
df = pd.read_csv("C:\\Users\\HP\\Downloads\\H1N1_Flu_Vaccines.csv")
```

```python
df
```

```python
df.drop("respondent_id",axis=1,inplace=True)
```

```python
df.dtypes
```

```python
float_pipeline = []
object_pipeline = []
index = 0
for _types in df.dtypes:
    if _types == "object":
        object_pipeline.append(df.columns[index])
        pass
    else:
        float_pipeline.append(df.columns[index])
        pass
    index = index + 1
    pass
```

```python
float_pipeline
```

```python
object_pipeline
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

num_pipline = Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    ('std_scaler',StandardScaler())
])

full_pipeline = ColumnTransformer([
    ("num",num_pipline,float_pipeline),
    ("cat",OneHotEncoder(),object_pipeline)
])

```

```python
prep = full_pipeline.fit_transform(df)
```

```python
label = prep[:,-2:]
data = prep[:,:-2]
```

```python
label
```

```python
from sklearn.neighbors import KNeighborsClassifier
sgd_clf = KNeighborsClassifier()
sgd_clf.fit(data,label)
```
