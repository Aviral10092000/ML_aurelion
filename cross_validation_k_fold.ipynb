```python
dataset = df[["Income","Purchased Bike"]]
dataset
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer


num_pipeline = Pipeline({
    ('imputer',SimpleImputer(strategy="median")),
    ('std_scaler',StandardScaler())
})

target_pipeline = Pipeline({
    ('imputer',SimpleImputer(strategy="most_frequent"))
})

```

```python
X = num_pipeline.fit_transform(dataset["Income"].to_numpy().reshape(-1,1))
y = target_pipeline.fit_transform(dataset["Purchased Bike"].to_numpy().reshape(-1,1))
```

```python
empty = []
for i in X:
    empty.append(i[0])
    pass

X = empty
empty = []
for i in y:
    empty.append(i[0])
    pass
y = empty
y
```

```python
club = pd.DataFrame({'X': X, 'y': y}, columns=['X', 'y'])
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(club, test_size=0.2, random_state=42)
```

```python
train_set["y"]
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.tree import DecisionTreeClassifier
scores = cross_val_score(DecisionTreeClassifier(), train_set["X"].to_numpy().reshape(-1,1), train_set["y"].to_numpy().reshape(-1,1), cv=3)
tree_rmse_scores = np.sqrt(scores)
```

```python
def display_result(scores):
    print("Scores:", scores)
    print("mean: ", scores.mean())
    print("standard deviation", scores.std())
    pass
display_result(tree_rmse_scores)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = [
    {'n_estimators':[3,10,30], 'max_features': [2,4,6,8]},
    {'bootstrap' : [False],'n_estimators':[3,10,30], 'max_features': [2,3,4]},
]

forest_class = DecisionTreeClassifier()
grid_search = GridSearchCV(forest_class,param_grid, cv=5,return_train_score=True)

grid_search.fit(X,y)
```
