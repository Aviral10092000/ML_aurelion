```python
import pandas as pd
import numpy as np

X = 2 * np.random.rand(100,1)
y = 4 + 3 * X + np.random.rand(100,1)
```

```python
from sklearn.linear_model import Lasso
import matplotlib.pyplot as plt
value = [0,.1,.5]
color = ["b-","r-","c-"]
_ = 0
for i in value:
    Lasso_reg = Lasso(alpha=i)
    Lasso_reg.fit(X,y)
    Lasso_reg.predict([[1.5]])
    plt.scatter(X,y)
    X_new = np.array([[0],[2]])
    y_predict = Lasso_reg.predict(X_new)
    plt.plot(X_new,y_predict,color[_])
    _ = _ +1
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
m = 100
X = 6*np.random.rand(m,1) - 3
y = 0.5 * X**2 + 2 + np.random.randn(m,1)
value = [0,1e-7,.5]
color = ["b-","y-","c-"]
_ = 0
plt.scatter(X,y)

for i in value:
    poly_features = PolynomialFeatures(degree = 10,include_bias=False)
    X_poly = poly_features.fit_transform(X)
    lasso_reg = Lasso(alpha=i)
    lasso_reg.fit(X_poly,y)
    _X = np.linspace(-3,3,1000)
    _y = [np.zeros(1000)]
    for i in range(10):
        _y = _y + (_X**(i+1))*lasso_reg.coef_[i]
        pass
    _y = _y + lasso_reg.intercept_

    plt.plot(_X,_y.ravel(),color[_])
    plt.scatter(X,y)
    _ = _ +1
    pass
plt.show()
```
